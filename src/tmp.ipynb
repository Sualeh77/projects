````python
# Initialize different SegFormer models with various encoders
encoders = ["mit_b0", "mit_b1", "mit_b2", "mit_b3", "mit_b4", "mit_b5"]

for encoder in encoders:
    model = smp.Segformer(
        encoder_name=encoder,
        encoder_weights=None,
        encoder_depth=5,
        in_channels=3,
        classes=3,
        decoder_channels=[256, 128, 64, 32, 16],
        decoder_use_batchnorm=True,
        decoder_attention_type=None
    )
    
    # Calculate total parameters
    total_params = sum(p.numel() for p in model.parameters())
    
    print(f"\nSegFormer with {encoder} encoder:")
    print(f"Total parameters: {total_params:,}")

````

````python
# Initialize different UNet++ models with various encoders
encoders = ["resnet18", "resnet34", "resnet50", "resnet101", "resnet152"]

for encoder in encoders:
    model = smp.UnetPlusPlus(
        encoder_name=encoder,
        encoder_weights=None,
        in_channels=3,
        classes=4,
        decoder_channels=(256, 128, 64, 32, 16),
        decoder_attention_type=None
    )
    
    # Calculate total parameters
    total_params = sum(p.numel() for p in model.parameters())
    
    print(f"\nUNet++ with {encoder} encoder:")
    print(f"Total parameters: {total_params:,}")
````

````python
import segmentation_models_pytorch as smp
# Test LR finder function
import torch
import torch.nn as nn
from src.utils.utils import find_optimal_lr

# Create a simple model
model = smp.Unet(
    encoder_name="resnet18",
    encoder_weights=None, 
    in_channels=3,
    classes=3
)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-7)

# Create dummy dataset and dataloader for testing
class DummyDataset:
    def __init__(self):
        self.data = [(torch.randn(3, 512, 512), torch.randint(0, 3, (512, 512))) for _ in range(100)]
    def __len__(self):
        return len(self.data)
    def __getitem__(self, idx):
        return self.data[idx]

train_loader = torch.utils.data.DataLoader(
    DummyDataset(),
    batch_size=4,
    shuffle=True
)

# Run LR finder
device = torch.device("mps")
model = model.to(device)

lr_finder = find_optimal_lr(
    model=model,
    train_loader=train_loader,
    criterion=criterion,
    optimizer=optimizer,
    device=device,
    min_lr=1e-7,
    max_lr=1,
    num_iter=100,
    log_dir="logs/lr_finder_test"
)

# Get suggested learning rate
# history = lr_finder.history
# lrs = [x["lr"] for x in history]
# losses = [x["loss"] for x in history]

# # Find point with steepest gradient
# gradients = np.gradient(losses)
# steepest_point = np.argmin(gradients)
# suggested_lr = lrs[steepest_point]

# print(f"Suggested learning rate: {suggested_lr:.1e}")
````
